# Bibliotecas necessárias

```python
import sys
sys.path.append('/home/jovyan/work')
from datetime import datetime, timedelta
from crawler_sti import create_df
from utils.SendBuckets import SendBuckets

tabela = 'sti_relatorios_sgt'
endpoint = 'https://apisgt.cni.org.br:8253/reports/v1.0.0/producoesMetrologia'


import time
import pandas as pd

# Defina o número de dias que deseja iterar, incluindo o dia atual
num_dias = 5

# Data atual menos o número de dias, incluindo o dia atual
start_date = datetime.now() - timedelta(days=num_dias - 1)
```

```python
#Bucket a ser gravado
send_buckets = SendBuckets('bronze')
```

# Extração dos dados

```python
# Lista para armazenar os DataFrames
dfs = []

# Loop sobre os dias
for i in range(num_dias):
    # Calcula a data atual do loop
    current_date = start_date + timedelta(days=i)
    # Converte a data atual para o formato desejado ('YYYYMMDD')
    date_str = current_date.strftime('%Y%m%d')
    # Chama a função create_df() com a data atual
    df = create_df(endpoint, date_str, tabela)
    # Append do DataFrame à lista
    dfs.append(df)
    
    # Pausa de 10 segundos
    time.sleep(10)


# Concatenar os DataFrames da lista
final_df = pd.concat(dfs, ignore_index=True)
```

```python
send_buckets.send_dataframe_bucket_part('overwrite', f'sti/apisgt/{tabela}', final_df, 'date_partition')
```
